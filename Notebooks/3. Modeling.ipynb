```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.model_selection import train_test_split
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'C:\Users\pedro\OneDrive\Desktop\Fraudulent-Actions-Prediction\Data\processed_data\train_ready_for_modeling.csv')
```

```python
df
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

### Perform Kfold Cross Validation (Balanced Dataset) ###
---

```python
from sklearn.model_selection import KFold, cross_validate
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
lr = LogisticRegression()
```

```python
lr_scores = cross_validate(lr, X, y, cv=kf, scoring=['accuracy', 'precision', 'recall', 'f1', 'roc_auc'])
```

```python
print ("Logistic Regression Scores:")
print("Accuracy:", lr_scores['test_accuracy'].mean())
print("Precision:", lr_scores['test_precision'].mean())
print("Recall:", lr_scores['test_recall'].mean())
print("F1 Score:", lr_scores['test_f1'].mean())
print("ROC AUC:", lr_scores['test_roc_auc'].mean())
```

```python
gb = GradientBoostingClassifier() #default is 100 trees per fold, 5 fold means 500 trees built. Took a long time to run (2+ hours)
gb_scores = cross_validate(gb, X, y, cv=kf, scoring=['accuracy', 'precision', 'recall', 'f1', 'roc_auc'])
```

```python
print("\nGradient Boosting Classifier Scores:")
print("Accuracy:", gb_scores['test_accuracy'].mean())
print("Precision:", gb_scores['test_precision'].mean())
print("Recall:", gb_scores['test_recall'].mean())
print("F1 Score:", gb_scores['test_f1'].mean())
print("ROC AUC:", gb_scores['test_roc_auc'].mean())
```

```python
nn = MLPClassifier(max_iter=200)
nn_scores = cross_validate(nn, X, y, cv=kf, scoring=['accuracy', 'precision', 'recall', 'f1', 'roc_auc'])
```

```python
print('\nNeural Network Classifier Scores:')
print("Accuracy:", nn_scores['test_accuracy'].mean())
print("Precision:", nn_scores['test_precision'].mean())
print("Recall:", nn_scores['test_recall'].mean())
print("F1 Score:", nn_scores['test_f1'].mean())
print("ROC AUC:", nn_scores['test_roc_auc'].mean()) 
```

As the dataset is well balanced, Accuracy might be the best metric to maximize. Otherwise, in a more 'real-world' scenarion, Recall would probably be the best metric to maximize.

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform
```

```python
param_dist = { 'C' : uniform(0.01,10),
               'solver' : ['lbfgs', 'liblinear'],
               'penalty' : ['l2', 'l1']
        }
```

```python
search = RandomizedSearchCV(estimator=lr, param_distributions=param_dist, n_iter=50, scoring='accuracy', cv=kf, random_state=42, n_jobs=-1)
```

```python
search.fit (X, y)
```

```python
print("Best params:", search.best_params_)
print("Best recall:", search.best_score_)
```

```python
final_model_lr = LogisticRegression(**search.best_params_)
final_model_lr.fit(X, y)
```
